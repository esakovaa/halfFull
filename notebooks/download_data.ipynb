# HalfFull — NHANES Raw Data Download

Downloads all NHANES 2017–March 2020 Pre-Pandemic `.xpt` files needed for the HalfFull project.

**Data source:** [CDC NHANES 2017–2020 Pre-Pandemic](https://wwwn.cdc.gov/nchs/nhanes/continuousnhanes/default.aspx?Cycle=2017-2020)

| Category | Search Page |
|----------|-------------|
| Demographics | https://wwwn.cdc.gov/nchs/nhanes/search/datapage.aspx?Component=Demographics&Cycle=2017-2020 |
| Dietary | https://wwwn.cdc.gov/nchs/nhanes/search/datapage.aspx?Component=Dietary&Cycle=2017-2020 |
| Examination | https://wwwn.cdc.gov/nchs/nhanes/search/datapage.aspx?Component=Examination&Cycle=2017-2020 |
| Laboratory | https://wwwn.cdc.gov/nchs/nhanes/search/datapage.aspx?Component=Laboratory&Cycle=2017-2020 |

> Files are saved to `data/raw/nhanes/` (gitignored). Run this notebook once before running any analysis notebooks.

```python
import urllib.request
from pathlib import Path

REPO_ROOT = Path('..').resolve()
OUT_DIR = REPO_ROOT / 'data' / 'raw' / 'nhanes'
OUT_DIR.mkdir(parents=True, exist_ok=True)

BASE_URL = 'https://wwwn.cdc.gov/nchs/nhanes/2017-2020'

print(f'Saving files to: {OUT_DIR}')
```

---
## 1. Demographics

```python
demographics_files = [
    ('P_DEMO.XPT', 'Demographics — Age, gender, ethnicity, income, survey weights'),
]

for filename, description in demographics_files:
    dest = OUT_DIR / filename.lower()
    if dest.exists():
        print(f'  SKIP  {filename} (already exists)')
    else:
        print(f'  GET   {filename} — {description}')
        urllib.request.urlretrieve(f'{BASE_URL}/{filename}', dest)
        print(f'        saved to {dest}')
```

## Summary
